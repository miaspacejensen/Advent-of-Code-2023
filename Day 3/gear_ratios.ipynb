```python
import re
```

```python
# read all lines in text file
with open("input.txt", "r") as file:
    lines = file.readlines()

# remove newline text from each line
lines = [line.rstrip("\n") for line in lines]

# get metadata
n_rows = len(lines)
n_cols = len(lines[0])
print(f"Shape: {n_cols} columns, {n_rows} rows")
```

```python
grid = [[' ' for _ in range(n_rows)] for _ in range(n_cols)]
```

```python
for i, line in enumerate(lines):
    for j, value in enumerate(line):
        grid[i][j] = value

# for row in grid:
#     print(''.join(row))
```

```python
offset_positions = [
    (-1, -1),
    (-1, 0),
    (-1, 1),
    (0, -1),
    (0, 0),
    (0, 1),
    (1, -1),
    (1, 0),
    (1, 1)
] 
```

```python
# lines2 = lines[:85]
# lines2
```

```python
all_part_numbers = []
all_part_numbers_sum = 0

for i, line in enumerate(lines):

    parts = re.split(r'\D+', line)
    parts = [part for part in parts if part]

    part_numbers = []
    
    for part in parts:
        
        start_j = line.index(part) 
        end_j = line.index(part) + len(part) - 1
        
        coords = []
        for j in range(start_j, end_j + 1):
            coords.append((i, j))
        
        # print(f"{part}, {coords}")

        statuses = []
        for j in range(start_j, end_j + 1):
            for offset_i, offset_j in offset_positions:
                if i in range(n_rows) and j in range(n_cols) and i + offset_i >= 0 and j + offset_j >= 0 and i + offset_i < n_rows and j + offset_j < n_cols:
                    offset_value = grid[i + offset_i][j + offset_j]
                    status = bool(re.search(r'[^0-9\.]', offset_value))
                    # print(j, grid[i + offset_i][j + offset_j], status)
                    statuses.append(status)

        if True in statuses:
            # print(f"(row {i}) -------- {part} is a part number")
            part_numbers.append(int(part))

        else:
            # print(f"(row {i}) -------- {part} is not")
            pass
            
        
    all_part_numbers.append(part_numbers)
    part_numbers_sum = sum(part_numbers)
    all_part_numbers_sum += part_numbers_sum


all_part_numbers = [part_numbers for part_numbers in all_part_numbers if part_numbers]
for i, nums in enumerate(all_part_numbers):
    print(f"(Row {i+1}) {nums}")
```
