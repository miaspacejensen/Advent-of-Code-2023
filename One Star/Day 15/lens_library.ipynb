#### Part 1

```python
# read all lines in text file
with open("input.txt", "r") as file:
    text = file.read() 

lines = text.split(",")
# lines
```

```python
def hash(line):
    current_value = 0
    for char in line:
        ascii = ord(char)
        current_value += ascii
        current_value *= 17
        current_value %= 256
    return current_value
```

```python
values = []

for line in lines:
    current_value = hash(line)
    values.append(current_value)
    
sum(values)

# correct answer = 508552
```

#### Part 2

```python
# read all lines in text file
with open("input.txt", "r") as file:
    text = file.read() 

lines = text.split(",")
# lines
```

```python
def check_if_lens_exists(item, dic):
    for box, lenses in dic.items():
        for i, lens in enumerate(lenses):
            if item in lens:
                return [box, lens, i]
    return False
```

```python
def replace_lens(new_line, old_line, dic, key, val):

    for i, item in enumerate(val):
        if old_line in val:
            dic[key][i] = new_line

    return dic
```

```python
n_boxes = 255
box_keys = list(range(n_boxes+1))
box_dic = dict(zip(box_keys, [[]]*len(box_keys)))
display_output = False

if display_output == True: 
    print(box_dic, "\n")

for i, line in enumerate(lines):

    if display_output == True: 
        print(f'({i}) After "{line}"')
    
    if "=" in line:
        line_new = line.replace('=', ' ')
        line_split = line.split("=")
        lens = line_split[0]
        focal_length = line_split[1]
        box = hash(lens)

        exists = check_if_lens_exists(lens, box_dic)
        if exists != False: 
            key_box = exists[0]
            val_lens = exists[1]
            val_index = exists[2]
            if display_output == True: 
                print(f">>> replace '{val_lens}' ({val_index}) from box {key_box}")
            box_dic[key_box][val_index] = line_new
        else:
            box_contents = box_dic[box] + [line_new]
            box_dic[box] = box_contents
        
    elif "-" in line:
        line_split = line.split("-")
        lens = line_split[0]
        focal_length = line_split[1]

        exists = check_if_lens_exists(lens, box_dic)
        if exists != False: 
            key_box = exists[0]
            val_lens = exists[1]
            val_index = exists[2]
            if display_output == True: 
                print(f">>> remove '{val_lens}' ({val_index}) from box {key_box}")
            del box_dic[key_box][val_index]
        
    else:
        pass
    
    if display_output == True: 
        print(box_dic)
        for key, val in box_dic.items():
            if len(val)>0:
                join_val = "["+"] [".join(val)+"]"
                print(f"Box {key}: {join_val}")

        print("")
```

```python
total_power = 0
for box in box_dic.keys():
    for i, lens in enumerate(box_dic[box]):
        lens_label = lens.split(" ")[0]
        focal_length = lens.split(" ")[1]
        power = (box + 1) * (i + 1) * int(focal_length)
        print(f"{lens_label}:     \t{box + 1} (box {box}) * {i + 1} (slot {i}) * {focal_length} (focal length) = {power}")
        total_power += power

print("\nTotal power =", total_power)

# 200000 - too low 
# 266675 - too high 
```
