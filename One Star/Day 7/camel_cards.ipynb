#### Part 1

```python
# read all lines in text file
with open("test_input.txt", "r") as file:
    lines = file.readlines() 
    
# remove newline text from each line
lines = [line.rstrip("\n").split(" ") for line in lines]
```

```python
cards = {
    "A": 14, 
    "K": 13,
    "Q": 12,
    "J": 11,
    "T": 10,
    "9": 9,
    "8": 8,
    "7": 7,
    "6": 6,
    "5": 5,
    "4": 4,
    "3": 3,
    "2": 2
}
```

```python
def five_of_a_kind(hand_dic):
    if 5 in hand_dic.values():
        return True

def four_of_a_kind(hand_dic):
    if 4 in hand_dic.values():
        return True

def full_house(hand_dic):
    if 2 in hand_dic.values() and 3 in hand_dic.values():
        return True

def three_of_a_kind(hand_dic):
    if 3 in hand_dic.values():
        return True

def two_pair(hand_dic):
    pair_count = 0
    for val in hand_dic.values():
        if val == 2:
            pair_count += 1
    if pair_count == 2:
        return True

def one_pair(hand_dic):
    pair_count = 0
    for val in hand_dic.values():
        if val == 2:
            pair_count += 1
    if pair_count == 1:
        return True
```

```python
rank = {}
lines_dic = {}

for line in lines:
    hand = line[0]
    bid = line[1]
    
    lines_dic[hand] = bid

    hand_dic = {}
    points = []
    
    for card in hand:
        points.append(cards[card])
        try:
            hand_dic[card] += 1
        except KeyError:
            hand_dic[card] = 1
    
    if five_of_a_kind(hand_dic):
        order = 7
    elif four_of_a_kind(hand_dic):
        order = 6
    elif full_house(hand_dic):
        order = 5
    elif three_of_a_kind(hand_dic):
        order = 4
    elif two_pair(hand_dic):
        order = 3
    elif one_pair(hand_dic):
        order = 2
    else:
        order = 1

    rank[hand] = (order, points)

sorted_rank = dict(sorted(rank.items(), key=lambda x: (x[1][0], x[1][1])))
sorted_rank
```

```python
result = 0
for i, key in enumerate(sorted_rank.keys()):
    rank = i+1
    bid = int(lines_dic[key])
    mult = rank*bid
    result += mult
    # print(rank, key, bid, mult)

result
```
